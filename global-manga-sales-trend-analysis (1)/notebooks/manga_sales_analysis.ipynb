# Global Manga Sales Trend Analysis

This notebook analyzes manga sales using Python libraries like Pandas and Seaborn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example data frame structure (replace with actual CSV later)
df = pd.DataFrame({
    'Title': ['One Piece', 'Naruto', 'Demon Slayer'],
    'Sales': [500, 370, 450]
})
sns.barplot(x='Title', y='Sales', data=df)
plt.title('Top Selling Manga')
plt.show()
```
